# Sudo

```python
%%sql
with total as (select count(*) as total from alias)
select 
    count(distinct alias_id) as '#', 
    round(count(distinct alias_id)*100.0/total,2) as '%' 
from alias join command using (alias_id)
join total
where command.sudo = 1
and command.name not like '-%'
```

```python
%%sql
with cmd_total as (select name, count(*) as cmd_total from command group by name)
select name, count(*) as '#', round(count(*)*100.0/cmd_total,2) as '%'
from command join cmd_total using (name)
where sudo = 1
and name not like '-%'
group by name
order by count(*) desc
limit 50;
```

```python
%%sql
with cmd_total as (select name, count(*) as cmd_total from command group by name)
select name, count(*) as num, round(count(*)*100.0/cmd_total,2) as per
from command join cmd_total using (name)
where sudo = 1
and name not like '-%'
group by name
having num > 100
order by 3 desc, 2 desc
limit 50;
```

```python
%%sql
with cmd_total as (select name, count(*) as cmd_total from command group by name)
select name, cmd_total, count(*) as num_sudo, round(count(*)*100.0/cmd_total,2) as per_sudo
from command join cmd_total using (name)
where sudo = 1
and name in ('brew','apt-get','apt','apt-cache','aptitude','$apt_pref','pacman','/usr/bin/pacman','yaourt','abs','pacaur','aur','zypper','yum','dnf','port','npm','gem','pip','pip3')
group by name
order by per_sudo desc, num_sudo desc;
```

```python
%%sql
with
    package_managers as (
        select * from command 
        where name in ('apt-get','apt','apt-cache','aptitude','$apt_pref','pacman','/usr/bin/pacman','abs','aur','zypper','yum','dnf','port','gem')
    ),
    total as (select count(*) as total from package_managers)
select round(count(*)*100.0/total,2) as '% of package managers'
from package_managers join total
where sudo = 1
```

```python
%%sql
with total_sudo as (select count(*) as total_sudo from command where sudo = 1)
select count(*), round(count(*)*100.0/total_sudo,2) as '% of sudo'
from command join total_sudo
where name in ('apt-get','apt','apt-cache','aptitude','$apt_pref','pacman','/usr/bin/pacman','abs','aur','zypper','yum','dnf','port','gem')
and sudo = 1
```
