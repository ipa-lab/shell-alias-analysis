# Environment Variables

How many aliases set environment variables?

```python
%%sql
with total as (select count(*) as total from alias)
select count(distinct alias_id) as '#', round(count(distinct alias_id)*100.0/total,2) as '%'
from alias join command using (alias_id)
join total
where command.name like '%=%';
```

```python
%%sql
with total as (select count(*) as total from alias)
select count(distinct alias_id) as '#', round(count(distinct alias_id)*100.0/total,2) as '%'
from alias join command using (alias_id)
join total
where command.name is 'export'
```

How often are certain commands called with environment vars being set beforehand?

```python
%%sql
with
    cmds as (
        select name, count(*) as total
        from command
        where name not like '-%'
        and name not like '%=%'
        group by name
    )
select 
    cmds.name,
    command.name || ' ' || command.arguments,
    cmds.total as '#', 
    count(*) as '# env', 
    round(count(*)*100.0/cmds.total,2) as per_env
from command join argument using (command_id)
join cmds
where command.name like '%=%'
and cmds.name = argument.name
and argument.position = 0
group by cmds.name
order by count(*) desc
limit 50;
```
